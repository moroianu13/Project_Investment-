# Calculating Future Value
The formula to calculate the future value of an investment is:
FV = PV x (1 + r)^n 

Where:
- **FV** is the future value of the investment.
- **PV** is the present value of the investment (10,000).
- **r** is the annual growth rate.
- **n** is the number of years (5 years).


```python
import numpy as np
import matplotlib.pyplot as plt

# Initial investment
initial_investment = 10000

# Growth rates (as decimals)
safe_countries_growth_rates = {
    'United States': 0.02,
    'Germany': 0.015,
    'Switzerland': 0.015,
    'Norway': 0.02,
    'Australia': 0.025
}

emerging_countries_growth_rates = {
    'India': 0.07,
    'Brazil': 0.025,
    'Indonesia': 0.055,
    'Mexico': 0.03,
    'Vietnam': 0.065
}

# Years
years = np.arange(0, 6)

# Function to calculate future value
def calculate_future_value(pv, rate, years):
    return pv * (1 + rate) ** years

# Calculate future values for each country
safe_countries_values = {country: calculate_future_value(initial_investment, rate, years) 
                         for country, rate in safe_countries_growth_rates.items()}

emerging_countries_values = {country: calculate_future_value(initial_investment, rate, years) 
                             for country, rate in emerging_countries_growth_rates.items()}

# Create DataFrames for plotting
safe_df = pd.DataFrame(safe_countries_values, index=years)
emerging_df = pd.DataFrame(emerging_countries_values, index=years)

# Plotting the growth of profit using bar and line plots
fig, ax = plt.subplots(2, 1, figsize=(14, 12))

# Plot for safe countries (bar + line)
safe_df.plot(kind='bar', ax=ax[0], alpha=0.7, width=0.7)
for country in safe_countries_values:
    ax[0].plot(years, safe_countries_values[country], marker='o', label=country)
ax[0].set_title('Growth of 10,000 Investment in Safe Countries Over 5 Years')
ax[0].set_xlabel('Years')
ax[0].set_ylabel('Investment Value ()')
ax[0].legend(loc='upper left')
ax[0].grid(True)

# Plot for emerging countries (bar + line)
emerging_df.plot(kind='bar', ax=ax[1], alpha=0.7, width=0.7)
for country in emerging_countries_values:
    ax[1].plot(years, emerging_countries_values[country], marker='o', label=country)
ax[1].set_title('Growth of 10,000 Investment in Emerging Countries Over 5 Years')
ax[1].set_xlabel('Years')
ax[1].set_ylabel('Investment Value ()')
ax[1].legend(loc='upper left')
ax[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Risk values (lower is safer)
safe_countries_risk = {
    'United States': 2,
    'Germany': 1,
    'Switzerland': 1,
    'Norway': 1,
    'Australia': 1
}

emerging_countries_risk = {
    'India': 10,
    'Brazil': 13,
    'Indonesia': 9,
    'Mexico': 8,
    'Vietnam': 12
}

# Combine the dictionaries for plotting
combined_risk = {**safe_countries_risk, **emerging_countries_risk}

# Create DataFrame for plotting
risk_df = pd.DataFrame.from_dict(combined_risk, orient='index', columns=['Risk'])

# Plotting the risk by country
plt.figure(figsize=(14, 8))
risk_df.sort_values(by='Risk', ascending=True).plot(kind='barh', legend=False)
plt.title('Investment Risk by Country')
plt.xlabel('Risk (Lower is Safer)')
plt.ylabel('Country')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from pptx import Presentation
from pptx.util import Inches
import pandas as pd
import os

# Function to save plots
def save_plot(fig, filename):
    fig.savefig(filename)
    plt.close(fig)

# Create the presentation object
prs = Presentation()

# Slide 1: Title Slide
slide = prs.slides.add_slide(prs.slide_layouts[0])
title = slide.shapes.title
subtitle = slide.placeholders[1]
title.text = "Investment Analysis of Emerging Markets"
subtitle.text = "Balancing Risk and Return for Optimal Investment Choices\nPrepared by: [Your Name]\nDate: [Date]"

# Slide 2: Introduction
slide = prs.slides.add_slide(prs.slide_layouts[1])
title, content = slide.shapes.title, slide.placeholders[1]
title.text = "Introduction"
content.text = ("Objective: To analyze and compare the investment potential of various emerging markets "
                "based on expected returns and associated risks.\n"
                "Key Metrics: Future value of a 10,000 investment over 5 years, credit ratings as risk indicators.")

# Slide 3: Data Assumptions
slide = prs.slides.add_slide(prs.slide_layouts[1])
title, content = slide.shapes.title, slide.placeholders[1]
title.text = "Data Assumptions"
content.text = ("We'll use the average GDP growth rates for the next 5 years:\n"
                "Safe Countries:\n"
                "  - United States: 2.0%\n"
                "  - Germany: 1.5%\n"
                "  - Switzerland: 1.5%\n"
                "  - Norway: 2.0%\n"
                "  - Australia: 2.5%\n\n"
                "Emerging Countries:\n"
                "  - India: 7.0%\n"
                "  - Brazil: 2.5%\n"
                "  - Indonesia: 5.5%\n"
                "  - Mexico: 3.0%\n"
                "  - Vietnam: 6.5%")

# Slide 4: Calculating Future Value
slide = prs.slides.add_slide(prs.slide_layouts[1])
title, content = slide.shapes.title, slide.placeholders[1]
title.text = "Calculating Future Value"
content.text = ("The formula to calculate the future value of an investment is:\n"
                "FV = PV  (1 + r)^n\n\n"
                "Where:\n"
                "- FV is the future value of the investment.\n"
                "- PV is the present value of the investment (10,000).\n"
                "- r is the annual growth rate.\n"
                "- n is the number of years (5 years).")

# Generate plots and save them as images
# Initial investment
initial_investment = 10000

# Growth rates (as decimals)
safe_countries_growth_rates = {
    'United States': 0.02,
    'Germany': 0.015,
    'Switzerland': 0.015,
    'Norway': 0.02,
    'Australia': 0.025
}

emerging_countries_growth_rates = {
    'India': 0.07,
    'Brazil': 0.025,
    'Indonesia': 0.055,
    'Mexico': 0.03,
    'Vietnam': 0.065
}

# Years
years = np.arange(0, 6)

# Function to calculate future value
def calculate_future_value(pv, rate, years):
    return pv * (1 + rate) ** years

# Calculate future values for each country
safe_countries_values = {country: calculate_future_value(initial_investment, rate, years) 
                         for country, rate in safe_countries_growth_rates.items()}

emerging_countries_values = {country: calculate_future_value(initial_investment, rate, years) 
                             for country, rate in emerging_countries_growth_rates.items()}

# Create DataFrames for plotting
safe_df = pd.DataFrame(safe_countries_values, index=years)
emerging_df = pd.DataFrame(emerging_countries_values, index=years)

# Plotting the growth of profit using bar and line plots
fig, ax = plt.subplots(figsize=(14, 8))

# Plot for safe countries (bar + line)
safe_df.plot(kind='bar', ax=ax, alpha=0.7, width=0.7)
for country in safe_countries_values:
    ax.plot(years, safe_countries_values[country], marker='o', label=country)

ax.set_title('Growth of 10,000 Investment in Safe Countries Over 5 Years')
ax.set_xlabel('Years')
ax.set_ylabel('Investment Value ()')
ax.legend(loc='upper left')
ax.grid(True)

# Save the plot
safe_plot_file = "safe_countries_plot.png"
save_plot(fig, safe_plot_file)

# Add plot to slide 5
slide = prs.slides.add_slide(prs.slide_layouts[5])
title = slide.shapes.title
title.text = "Growth of 10,000 Investment in Safe Countries Over 5 Years"
left = Inches(1)
top = Inches(2)
pic = slide.shapes.add_picture(safe_plot_file, left, top, width=Inches(8), height=Inches(4))

# Plot for emerging countries (bar + line)
fig, ax = plt.subplots(figsize=(14, 8))

emerging_df.plot(kind='bar', ax=ax, alpha=0.7, width=0.7)
for country in emerging_countries_values:
    ax.plot(years, emerging_countries_values[country], marker='o', label=country)

ax.set_title('Growth of 10,000 Investment in Emerging Countries Over 5 Years')
ax.set_xlabel('Years')
ax.set_ylabel('Investment Value ()')
ax.legend(loc='upper left')
ax.grid(True)

# Save the plot
emerging_plot_file = "emerging_countries_plot.png"
save_plot(fig, emerging_plot_file)

# Add plot to slide 6
slide = prs.slides.add_slide(prs.slide_layouts[5])
title = slide.shapes.title
title.text = "Growth of 10,000 Investment in Emerging Countries Over 5 Years"
left = Inches(1)
top = Inches(2)
pic = slide.shapes.add_picture(emerging_plot_file, left, top, width=Inches(8), height=Inches(4))

# Slide 7: Risk by Country
# Risk values (lower is safer)
safe_countries_risk = {
    'United States': 2,
    'Germany': 1,
    'Switzerland': 1,
    'Norway': 1,
    'Australia': 1
}

emerging_countries_risk = {
    'India': 10,
    'Brazil': 13,
    'Indonesia': 9,
    'Mexico': 8,
    'Vietnam': 12
}

# Combine the dictionaries for plotting
combined_risk = {**safe_countries_risk, **emerging_countries_risk}

# Create DataFrame for plotting
risk_df = pd.DataFrame.from_dict(combined_risk, orient='index', columns=['Risk'])

# Plotting the risk by country
fig, ax = plt.subplots(figsize=(14, 8))
risk_df.sort_values(by='Risk', ascending=True).plot(kind='barh', legend=False, ax=ax)
ax.set_title('Investment Risk by Country')
ax.set_xlabel('Risk (Lower is Safer)')
ax.set_ylabel('Country')
ax.grid(True)

# Save the plot
risk_plot_file = "risk_plot.png"
save_plot(fig, risk_plot_file)

# Add plot to slide 7
slide = prs.slides.add_slide(prs.slide_layouts[5])
title = slide.shapes.title
title.text = "Investment Risk by Country"
left = Inches(1)
top = Inches(2)
pic = slide.shapes.add_picture(risk_plot_file, left, top, width=Inches(8), height=Inches(4))

# Slide 8: Interpretation
slide = prs.slides.add_slide(prs.slide_layouts[1])
title, content = slide.shapes.title, slide.placeholders[1]
title.text = "Interpretation"
content.text = ("Safe Countries:\n"
                "- United States: Investment grows to approximately 11,041.\n"
                "- Germany: Investment grows to approximately 10,772.\n"
                "- Switzerland: Investment grows to approximately 10,772.\n"
                "- Norway: Investment grows to approximately 11,041.\n"
                "- Australia: Investment grows to approximately 11,331.\n\n"
                "Emerging Countries:\n"
                "- India: Investment grows to approximately 14,026.\n"
                "- Brazil: Investment grows to approximately 11,314.\n"
                "- Indonesia: Investment grows to approximately 13,071.\n"
                "- Mexico: Investment grows to approximately 11,593.\n"
                "- Vietnam: Investment grows to approximately 13,696.")

# Slide 9: Conclusion
slide = prs.slides.add_slide(prs.slide_layouts[1])
title, content = slide.shapes.title, slide.placeholders[1]
title.text = "Conclusion"
content.text = ("Best Choice for Investment:\n"
                "Based on the balance of risk and return, Indonesia emerges as the best choice for investment among the given emerging markets. It provides a high expected return with a moderate risk profile:\n"
                "-Expected Return: 13,071 from a 10,000 investment over 5 years.\n"
                "-Risk: Rated BBB (9 on our risk scale)." )               
```
